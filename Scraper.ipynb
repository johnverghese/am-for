```python
import requests
import urllib.request
from urllib.parse import quote
import time
from bs4 import BeautifulSoup
import csv
from geopy.geocoders import Nominatim
```

```python
trees = []
i = 1

while i < 30:

    url = 'https://www.americanforests.org/get-involved/americas-biggest-trees/champion-trees-national-register/?submit_search=Search&search_val=&order_by=&bt_page_id=' \
                + str(i)
    page = requests.get(url)

    soup = BeautifulSoup(page.content, 'html.parser')

    tree_elems = soup.find_all(class_='bigtrees_results_section')
    
    if not tree_elems:
        break
        
    for tree_elem in tree_elems:
        name = tree_elem.find(class_ = 'bigtrees_results_title').text
        link = tree_elem.find(class_ = 'tree_img_span').find('a')['href']
        img = tree_elem.find(class_ = 'tree_img_span').find('img')['src']
        sci_name, state, county, points = (info.text for info in tree_elem.find_all(class_ = 'bigtrees_results_ans'))
        tree = {'name': name, 'sci_name': sci_name, 'state': state, 'county': county, 'points': points, 'link': link, 'img': img}
        trees.append(tree)
        
    i += 1
    
    time.sleep(1)

trees
```
